```python
names=["pandu","nagarjuna","samreen","suresh","nithish"]
sals=range(10000,200000,1000)
import random
f=open("data.csv","w")
f.write("id,name,sal")
for i in range(100):
    row="\n%s,%s,%s"%(i,random.choice(names),random.choice(sals))
    f.write(row)
    f.flush()
f.close()
```
